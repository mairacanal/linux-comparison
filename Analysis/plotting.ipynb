```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv('../../Fapesp/Aquisitions/Pesquisa/Teste 4 - Kernelspace com stress/results/18000Hz.csv', float_precision='high')
df.head(1)
```

```python
delay = []
jitter = []

for d in df['delay']:
    delay.append(d)
        
for j in df['jitter']:
    jitter.append(j)
```

```python
print("Latência média:" + str(np.mean(delay)))
print("Latência máxima:" + str(np.max(delay)))
print("Desvio Padrão:" + str(np.std(delay)))
```

```python
# histograma de latência

density = stats.gaussian_kde(delay)
n, x, _ = plt.hist(delay, bins=50, 
                   histtype=u'step', density=True)  
plt.plot(x, density(x))
plt.xlabel('Latência (s)')
plt.ylabel('Número de amostras')
plt.show()
```

```python
# Variação da latência no tempo
plt.xlabel('Amostra')
plt.ylabel('Latência (s)')
plt.plot(df['req_id'], delay)
```

```python
# Variação do jitter no tempo
plt.xlabel('Amostra')
plt.ylabel('Jitter (s)')
plt.plot(df['req_id'], jitter)
```

```python
print("Jitter média:" + str(np.mean(jitter)))
print("Jitter máxima:" + str(np.max(jitter)))
print("Desvio Padrão:" + str(np.std(jitter)))
```
